## HackerEarth Machine Learning ...solving citizensâ€™ grievances

In this problem, we are given a dataset that contains grievances of various people living in a country. Our task is to predict the importance of the grievance with respect to various articles, constitutional declarations, enforcement, resources, and so on, to help the government prioritize which ones to deal with and when.

```python
#import libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


import warnings
warnings.filterwarnings('ignore')
```

```python
#Enablle maximum rows and column for dataset

pd.pandas.set_option('display.max_rows',None)
pd.pandas.set_option('display.max_columns',None)
```

<b> Load the data

```python
df = pd.read_csv('Train.csv') #train set
df_test = pd.read_csv('Test.csv') #test set
```

```python
print('Size of train data ----',df.shape)
print('Size of test data ----',df_test.shape)
```

```python
#lets show the head of dataset
df.head(2)
```

```python
#function to check null percentage 

def check_null(data):
    i=0
    for col in data.columns:
        if data[col].isnull().sum()>0:
            print(col,'-- Null values --', data[col].isnull().sum(),', percentage-', round((data[col].isnull().sum()/len(data))*100,2),'%')
            i=i+1
    print('Total no of columns ',i)
```

```python
# null value of train

check_null(df)
```

```python
# null value of test

check_null(df_test)
```

<b>Note :</b> We can delete all the columns of high missing values

```python
#drop all the columns of high missing values

df.drop(['decisiondate','introductiondate','issue.0','issue.1','issue.10','issue.11','issue.12','issue.13',
        'issue.14','issue.15','issue.16','issue.17','issue.18','issue.19','issue.2','issue.20','issue.21',
       'issue.22','issue.23','issue.24','issue.25','issue.26','issue.3','issue.4','issue.5','issue.6','issue.7','issue.8',
       'issue.9','parties.2','respondent.1','respondent.2','respondent.3','respondent.4'],axis=1,inplace=True)
```

```python
#test set
df_test.drop(['decisiondate','introductiondate','issue.0','issue.1','issue.10','issue.11','issue.12','issue.13',
        'issue.14','issue.15','issue.16','issue.17','issue.18','issue.19','issue.2','issue.20','issue.21',
       'issue.22','issue.23','issue.24','issue.25','issue.26','issue.3','issue.4','issue.5','issue.6','issue.7','issue.8',
       'issue.9','parties.2','respondent.1','respondent.2','respondent.3','respondent.4'],axis=1,inplace=True)
```

```python
#present dimension of data

print(df.shape)
print(df_test.shape)
```

```python
df.head(2)
```

```python
#create a function to check uniqueness in columns 
#we'll find out column with more than 70 unique value, i.e. high cardinality

def uniqueness (col,data):
    if data[col].nunique()/len(data) > 0.7:
        print('For column ', col, 'No of unique field ',data[col].nunique(),' % of unique= ', round(data[col].nunique()/len(data),2))
```

```python
#check the uniqueness of the columns

for col in df.columns:
    uniqueness(col,df)
```

```python
for col in df_test.columns:
    uniqueness(col,df_test)
```

<b>Note: </b> The above columns have high cardinality so may be dropped

```python
#drop the columns with high cardinality

for col in df.columns:
    if df[col].nunique()/len(df) > 0.7:
        df.drop(col,axis=1,inplace=True)
```

```python
# drop the same from test set

for col in df_test.columns:
    if df_test[col].nunique()/len(df_test) > 0.7:
        df_test.drop(col,axis=1,inplace=True)
```

```python
#present dimension of data

print(df.shape)
print(df_test.shape)
```

```python
#for test set

for col in df_test.columns:
    if df_test[col].nunique() == 1:
        print('column name=',col,'--single value in the column-- ',df_test[col].unique())
```

We'll drop columns which have single value in both dataset


```python
for col in df.columns:
    if df[col].nunique() == 1 and df_test[col].nunique() == 1:
        df.drop(col,axis=1, inplace=True)
        df_test.drop(col,axis=1, inplace=True)
```

```python
#present dimension of data

print(df.shape)
print(df_test.shape)
```

Also, we can observe <b>country.alpha2</b> and <b>country.name</b> are actually the same fields with country names and abbreviated format, so we'll going to keep only <b>country.name</b>

```python
df.drop('country.alpha2', axis=1, inplace=True)
df_test.drop('country.alpha2', axis=1, inplace=True)
```

```python
#Find our all the categorical columns

df.select_dtypes(include='object').columns
```

```python
#Find our all the boolean columns

df.select_dtypes(include='bool').columns
```

```python
#convert boolean to binary

df['separateopinion'] = df['separateopinion'].astype(int)
df_test['separateopinion'] = df_test['separateopinion'].astype(int)
```

For the time being we are going to drop 'kpdate', 'parties.1' columns

```python
df.drop(['kpdate','parties.1'],axis=1,inplace=True)
df_test.drop(['kpdate','parties.1'],axis=1,inplace=True)
```

```python
#present dimension of data

print(df.shape)
print(df_test.shape)
```

```python
#convert judgementdate column to year and month

df['judgementdate'] = pd.to_datetime(df['judgementdate'])
df['Year_jdate'] = df['judgementdate'].apply(lambda date : date.year)
df['Month_jdate'] = df['judgementdate'].apply(lambda date : date.month)

df_test['judgementdate'] = pd.to_datetime(df_test['judgementdate'])
df_test['Year_jdate'] = df_test['judgementdate'].apply(lambda date : date.year)
df_test['Month_jdate'] = df_test['judgementdate'].apply(lambda date : date.month)
```

```python
#drop judgementdate after transfering it

df.drop('judgementdate',axis=1,inplace=True)
df_test.drop('judgementdate',axis=1,inplace=True)
```

We'll create dummies on categorical columns

```python
for col in df.columns:
    if df[col].dtype == 'O':
        df = pd.get_dummies(df,columns=[col])
        df_test = pd.get_dummies(df_test,columns=[col])
```
